## Data Cleaning

```python
import pandas as pd
from IPython.display import display, Markdown

# Step 1: Filter out candidates
df_candidates = df[df['koi_disposition'] == 'CANDIDATE'].copy()
df_binary = df[df['koi_disposition'].isin(['CONFIRMED', 'FALSE POSITIVE'])].copy()

# Step 2: Drop fully or mostly useless columns
columns_to_remove = ['koi_teq_err1', 'koi_teq_err2', 'kepler_name']
df_binary.drop(columns=columns_to_remove, inplace=True)

# Document removals
columns_removed = [
    {'Column': 'koi_teq_err1', 'Reason': '100% missing'},
    {'Column': 'koi_teq_err2', 'Reason': '100% missing'},
    {'Column': 'kepler_name', 'Reason': '76% missing and not useful for modeling'},
]

# Step 3: Separate error columns and remove from main data
error_cols = [col for col in df_binary.columns if col.endswith('_err1') or col.endswith('_err2')]
df_errors = df_binary[error_cols].copy()
df_main = df_binary.drop(columns=error_cols)

for col in error_cols:
    columns_removed.append({'Column': col, 'Reason': 'Temporarily dropped (error margins)'})

# Step 4: Drop other non-useful columns
df_main.drop(columns=['koi_score', 'koi_tce_delivname'], inplace=True)
columns_removed += [
    {'Column': 'koi_score', 'Reason': '~15% missing, not core feature'},
    {'Column': 'koi_tce_delivname', 'Reason': 'Text column, irrelevant to modeling'}
]

# Step 5: Display clean summary table
removed_df = pd.DataFrame(columns_removed)
display(Markdown("### 🗂️ Columns Removed with Reasons"))
display(removed_df)

# Final check: shape of working dataset
print("✅ Final shape of working dataset:", df_main.shape)
```

```python
from IPython.display import display, Markdown

# Count total missing values
total_missing = df_main.isna().sum().sum()

if total_missing == 0:
    display(Markdown("✅ **No missing values found in `df_main`.**"))
else:
    display(Markdown(f"⚠️ **There are {int(total_missing)} missing values in `df_main`.**"))

    # Show top 10 columns with missing values
    display(Markdown("### 🔍 Top columns with missing values"))
    missing_summary = df_main.isna().sum()
    display(missing_summary[missing_summary > 0].sort_values(ascending=False).to_frame(name='Missing'))
```

```python
# Columns we want to fill with median values (numeric + relevant)
columns_to_fill = [
    'koi_depth', 'koi_prad', 'koi_teq', 'koi_model_snr',
    'koi_impact', 'koi_slogg', 'koi_steff', 'koi_srad',
    'koi_kepmag', 'koi_insol', 'koi_tce_plnt_num'
]

# Fill all of them with their respective median values
for col in columns_to_fill:
    median_value = df_main[col].median()
    df_main[col].fillna(median_value, inplace=True)
    print(f"✅ Filled missing values in '{col}' with median = {median_value:.2f}")
```

## Outliers

```python
# Exclude these columns from analysis
excluded_cols = ['label', 'kepid', 'rowid']

# Select numeric columns to analyze (exclude binary and ID)
numeric_cols = df_main.select_dtypes(include=['float64', 'int64']).columns
numeric_cols = [col for col in numeric_cols if col not in excluded_cols and df_main[col].nunique() > 10]

# Analyze outliers using IQR
outlier_summary = []

for col in numeric_cols:
    Q1 = df_main[col].quantile(0.25)
    Q3 = df_main[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df_main[(df_main[col] < lower) | (df_main[col] > upper)].shape[0]
    outlier_pct = outliers / df_main.shape[0] * 100
    outlier_summary.append((col, outliers, round(outlier_pct, 2)))

# Create and display results as a DataFrame
outlier_df = pd.DataFrame(outlier_summary, columns=["Column", "Outlier Count", "Outlier %"])
outlier_df = outlier_df.sort_values(by="Outlier %", ascending=False)

outlier_df
```

# Scaling & Label Creation

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Create binary label
if 'label' not in df_main.columns:
    df_main['label'] = df_main['koi_disposition'].map({'CONFIRMED': 1, 'FALSE POSITIVE': 0})
    df_main.drop(columns=['koi_disposition'], inplace=True)
    print("✅ Created binary `label` column from `koi_disposition`")

# Identify binary columns (do NOT scale them)
binary_cols = [col for col in df_main.columns
               if df_main[col].nunique() == 2 and set(df_main[col].unique()) <= {0, 1}]

# Select numeric columns to scale (excluding label, IDs, binary, and metadata columns)
excluded_from_scaling = binary_cols + ['label', 'kepid', 'rowid', 'ra', 'dec']
columns_to_scale = [col for col in df_main.select_dtypes(include=['float64', 'int64']).columns
                    if col not in excluded_from_scaling]

# Scale
scaler = StandardScaler()
df_main[columns_to_scale] = scaler.fit_transform(df_main[columns_to_scale])
print("📏 Scaled columns:", columns_to_scale)

# Save to CSV
df_main.to_csv("kepler_clean_scaled.csv", index=False)
print("✅ File saved to: kepler_clean_scaled.csv")
```
